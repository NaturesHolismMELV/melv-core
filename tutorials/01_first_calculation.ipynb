## Part 1: A Well-Documented Mutualism

### Cleaner Wrasse and Client Fish

One of the most thoroughly studied examples of cooperation in nature is the cleaner wrasse mutualism — the cleaner wrasse (*Labroides dimidiatus*) and its client fish.

**The relationship:**
- Cleaner wrasse remove parasites, dead tissue, and debris from client fish
- Client fish provide a food source (parasites) and cleaning station access
- Both species benefit significantly from the interaction

**Why does this cooperation emerge? Let's use MELV to find out.**

**Key observations:**
- **Resource overlap (ρ)**: VERY LOW (≈ 0.20)
  - Cleaners eat parasites, dead tissue, and mucus
  - Clients eat completely different prey (plankton, small fish, algae)
  - Almost no competition for food resources

- **Service differentiation (δ)**: HIGH (≈ 0.85)
  - Cleaners provide a critical health service (parasite removal can be life-or-death)
  - Clients provide a reliable food source and protection at cleaning stations
  - Strong mutual benefit through complementary services

Let's calculate the i-factor:

```python
# Calculate i-factor for cleaner wrasse mutualism
cleaner_result = melv.calculate_i_factor(
    overlap=0.20,
    differentiation=0.85
)

print("=== CLEANER WRASSE & CLIENT FISH MUTUALISM ===")
print(f"\nResource overlap (ρ):        {cleaner_result['overlap']:.2f}")
print(f"Service differentiation (δ): {cleaner_result['differentiation']:.2f}")
print(f"\ni-factor: {cleaner_result['i_factor']:.2f}")
print(f"Regime:   {cleaner_result['regime']}")
print(f"\nInterpretation:")
print(cleaner_result['interpretation'])
```

### Understanding the Result

**i = 0.24 < 1.0 → Cooperation regime**

Why?
- Very low overlap (0.20) means minimal resource competition
- High differentiation (0.85) means strong mutual benefit
- The ratio i = 0.20 / 0.85 = 0.24 is well below the critical threshold of 1.0
- Therefore: cooperation is energetically favourable

**This isn't altruism — it's mathematics.** The energetic cost of interaction is much less than the benefit received, making cooperation the optimal strategy.

> **Field grounding:** Hornbill-mongoose mutualism (Namibia, 1981–1983) and bee-flower energetic associations provided the original empirical anchors for the MELV framework. Both involve complementary services with low resource overlap — structurally identical to the cleaner wrasse case above. See Evans (1999, 2026).

**MELV successfully predicts this well-documented mutualism.**

## Part 2: The Modern Example

### Social Media Platforms

**Observations:**
- **Resource overlap (ρ)**: HIGH (≈ 0.80)
  - Competing for the same users
  - Same advertisers
  - Same attention economy
  - Similar content creators

- **Service differentiation (δ)**: MODERATE (≈ 0.50)
  - Some unique features (Stories, Reels, Spaces)
  - Different user experiences
  - But largely similar value propositions
  - Core services overlap significantly

What does MELV predict?

```python
# Calculate i-factor for social media platforms
social_result = melv.calculate_i_factor(
    overlap=0.80,
    differentiation=0.50
)

print("=== SOCIAL MEDIA PLATFORMS ===")
print(f"\nResource overlap (ρ):        {social_result['overlap']:.2f}")
print(f"Service differentiation (δ): {social_result['differentiation']:.2f}")
print(f"\ni-factor: {social_result['i_factor']:.2f}")
print(f"Regime:   {social_result['regime']}")
print(f"\nInterpretation:")
print(social_result['interpretation'])
```

### Understanding the Result

**i = 1.60 > 1.0 → Competition regime**

Why?
- High overlap (0.80) creates intense resource competition
- Moderate differentiation (0.50) provides insufficient mutual benefit
- The ratio i = 0.80 / 0.50 = 1.60 exceeds the critical threshold
- Therefore: competition is energetically dominant

**This explains:**
- Why platforms rarely cooperate (integrate, share users)
- Why we see "walled gardens" and data silos
- Why zero-sum dynamics dominate social media

**The same mathematical structure that explains cleaner wrasse explains tech platforms.**

> **Relationship to the full model:** In the 2026 master equation, high resource overlap corresponds to high ε (adaptation pressure) and low β (compatibility) — the modulation term (1 − ε × φ × β) remains large, keeping the system away from the cooperative attractor near i → 1.

## Part 3: The Critical Threshold

### The Bifurcation Point

At i ≈ 1.0, a sharp transition occurs between cooperation and competition regimes.

> **Version note:** In this v1 simplified model, the bifurcation occurs at exactly i = 1.0 (where ρ = δ). In the full 2026 master equation, full cooperative convergence (CI → 1.0) occurs asymptotically as i approaches its theoretical maximum near unity — a mechanistically richer but qualitatively consistent prediction.

Let's explore the threshold region:

```python
# Calculate i-factors across a range of overlap values
# Fixed differentiation at 0.50
overlaps = np.linspace(0.1, 0.9, 50)
diff = 0.50
i_factors = overlaps / diff

# Create visualisation
plt.figure(figsize=(12, 6))

# Plot i-factor curve
plt.plot(overlaps, i_factors, 'b-', linewidth=2, label='i-factor')

# Mark critical threshold
plt.axhline(y=1.0, color='r', linestyle='--', linewidth=2,
            label='Critical threshold (i = 1.0)')

# Shade cooperation and competition regions
plt.axhspan(0, 1.0, alpha=0.2, color='green', label='Cooperation regime')
plt.axhspan(1.0, 2.0, alpha=0.2, color='red', label='Competition regime')

# Mark our examples
plt.scatter([0.20], [0.20/0.85], s=200, c='darkgreen', marker='o',
            edgecolors='black', linewidths=2,
            label='Cleaner Wrasse (i = 0.24)', zorder=5)
plt.scatter([0.80], [0.80/0.50], s=200, c='darkred', marker='s',
            edgecolors='black', linewidths=2,
            label='Social Media (i = 1.60)', zorder=5)

plt.xlabel('Resource Overlap (ρ)', fontsize=12)
plt.ylabel('i-factor', fontsize=12)
plt.title('MELV v1 Bifurcation: Cooperation vs. Competition\n'
          '(Differentiation δ = 0.50)', fontsize=14, fontweight='bold')
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.ylim(0, 2.0)
plt.tight_layout()
plt.show()

print("\n=== KEY INSIGHT ===")
print("The critical threshold at i = 1.0 creates a sharp bifurcation:")
print("  • Below: Cooperation is energetically favourable")
print("  • Above: Competition dominates")
print("  • At threshold: System is highly sensitive to small changes")
print("\nThis bifurcation is a simplified analogue of the near-unity")
print("bifurcation in the full 2026 master equation.")
```

## Part 4: Uncertainty Quantification

### Real-World Measurements Have Uncertainty

In practice, we rarely know exact values for overlap and differentiation. Let's quantify this uncertainty using bootstrap confidence intervals:

```python
# Recalculate cleaner wrasse with uncertainty
cleaner_uncertain = melv.calculate_i_factor(
    overlap=0.20,
    differentiation=0.85,
    uncertainty=0.05,   # ±5% measurement error
    bootstrap_n=1000,
    random_state=42     # For reproducibility
)

ci_lower, ci_upper = cleaner_uncertain['confidence_interval']

print("=== CLEANER WRASSE WITH UNCERTAINTY ===")
print(f"\nPoint estimate:          i = {cleaner_uncertain['i_factor']:.2f}")
print(f"95% Confidence Interval:     [{ci_lower:.2f}, {ci_upper:.2f}]")
print(f"\nConclusion: Even with ±5% measurement error, we can confidently")
print(f"say that i < 1.0, predicting the COOPERATION regime.")

# Visualise the uncertainty
plt.figure(figsize=(10, 6))
x_pos = np.array([1, 2])
i_values = np.array([cleaner_uncertain['i_factor'], social_result['i_factor']])
errors = np.array([
    [cleaner_uncertain['i_factor'] - ci_lower,
     ci_upper - cleaner_uncertain['i_factor']],
    [0, 0]  # No uncertainty shown for social media example
]).T

plt.bar(x_pos, i_values, yerr=errors, capsize=10,
        color=['green', 'red'], alpha=0.7,
        edgecolor='black', linewidth=2)
plt.axhline(y=1.0, color='black', linestyle='--',
            linewidth=2, label='Critical threshold')
plt.xticks(x_pos, ['Cleaner Wrasse\n(Labroides dimidiatus)',
                    'Social Media\nPlatforms'])
plt.ylabel('i-factor', fontsize=12)
plt.title('i-Factor Comparison with Uncertainty\n'
          '(Error bars show 95% CI from bootstrap)',
          fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.ylim(0, 2.0)
plt.tight_layout()
plt.show()
```

## Part 6: Regime Scanning

### How Do Changes Affect the Regime?

Let's create a 2D map showing how different combinations of overlap and differentiation determine the cooperation/competition regime:

```python
# Create meshgrid
overlap_range = np.linspace(0.1, 0.9, 50)
diff_range = np.linspace(0.1, 0.9, 50)
O, D = np.meshgrid(overlap_range, diff_range)

# Calculate i-factor for each combination
I = O / D

# Create heatmap
plt.figure(figsize=(12, 10))

contour = plt.contourf(O, D, I, levels=20, cmap='RdYlGn_r', alpha=0.8)
plt.colorbar(contour, label='i-factor')

# Add critical threshold contour line
plt.contour(O, D, I, levels=[1.0], colors='black', linewidths=3,
            linestyles='--')

# Mark ecological examples
plt.scatter([0.20], [0.85], s=300, c='white', marker='o',
            edgecolors='black', linewidths=3,
            label='Cleaner Wrasse (i = 0.24)', zorder=5)
plt.scatter([0.80], [0.50], s=300, c='white', marker='s',
            edgecolors='black', linewidths=3,
            label='Social Media (i = 1.60)', zorder=5)

plt.xlabel('Resource Overlap (ρ)', fontsize=14, fontweight='bold')
plt.ylabel('Service Differentiation (δ)', fontsize=14, fontweight='bold')
plt.title('MELV v1 Regime Map: When Does Cooperation Emerge?\n'
          'Green = Cooperation (i < 1)  |  Red = Competition (i > 1)  '
          '|  Black dashed = Critical threshold',
          fontsize=13, fontweight='bold')
plt.legend(loc='upper right', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n=== REGIME MAP INSIGHTS ===")
print("\n• Upper-left (green): High differentiation, low overlap → COOPERATION")
print("  Examples: Cleaner wrasse, hornbill-mongoose, bee-flower associations")
print("\n• Lower-right (red): Low differentiation, high overlap → COMPETITION")
print("  Examples: Social media platforms, direct market competitors")
print("\n• Diagonal (black dashed): Critical threshold where regime shifts")
print("  Small changes near threshold can flip cooperation ↔ competition")
```

## Summary: What You've Learned

### Key Concepts

1. **The i-factor (v1 simplified)** quantifies interaction tendency:
   - i = resource_overlap / service_differentiation
   - Simple ratio with clear ecological interpretation

2. **Critical threshold at i ≈ 1.0**:
   - Below: Cooperation regime (differentiation exceeds overlap)
   - Above: Competition regime (overlap exceeds differentiation)
   - Sharp bifurcation, not a gradual transition

3. **Universal applicability**:
   - Same mathematical structure for cleaner wrasse, tech platforms, ecosystems
   - Cooperation is energetic efficiency, not altruism

4. **Practical implications**:
   - To favour cooperation: ↓ overlap OR ↑ differentiation
   - Small changes near threshold have large effects
